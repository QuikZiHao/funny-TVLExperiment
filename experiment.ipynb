```python
import random

class person:
    def __init__(self,id:int,ability:float):
        self.id = id
        self.money = 10
        self.ability = ability

    def event(self)->str:
        flag = random.random()
        if flag <= 0.2:
            return "no event"
        if flag <= 0.6:
            self.greenEvent()
            return "green"
        self.redEvent()
        return "red"

    def greenEvent(self):
        flag = random.random()
        if flag <= self.ability:
            self.money = self.money * 2

    def redEvent(self):
         self.money = self.money / 2
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class experimentModule:
    def __init__(self,num_participate:int):
        self.num_participate = num_participate
        random_numbers = np.random.normal(loc=0.6, scale=0.1, size=num_participate)
        self.person_info = [person(i,random_numbers[i])for i in range(num_participate)]

    def iter(self):
        eventHash = {"red":0,"green":0,"no event":0}
        for i in self.person_info:
            eventHash[i.event()] += 1
        print(eventHash)
    
    def summary(self):
        money = [i.money for i in self.person_info]
        money = sorted(money)
        info_Hash = {
            "B40": money[0:int(self.num_participate*0.4)],
            "M40": money[int(self.num_participate*0.4):int(self.num_participate*0.8)],
            "T20":money[int(self.num_participate*0.8):]
        }
        for i in info_Hash:
            print("summary of ",i," Avg = ",sum(info_Hash[i])/len(info_Hash[i])," ,Median = ",np.median(info_Hash[i]))
    
    def showTop20(self,year):
        info = [{"year":year,"id":i.id,"ability":i.ability,"money":float(i.money)} for i in self.person_info]
        info = sorted(info,key=lambda x:x["money"])
        return pd.DataFrame(info[800:])
    
    def draw(top20Data:pd.DataFrame,year):
        plt.hist(top20Data['ability'], bins=5, edgecolor='black')
        plt.xlabel('ability')
        plt.ylabel('Frequency')
        plt.title(str(year)+'Histogram of ability')
        plt.grid(True)
        plt.show()


```

```python
model = experimentModule(1000)
year = 40
dataHash = {}
for i in range(1,year+1):
     for j in range(2):
          model.iter()
          model.summary()
          dataHash[i+j/2] = model.showTop20(i+j/2)
          experimentModule.draw(dataHash[i+j/2],i+j/2)
```

```python
df = dataHash[40.5]
df = df.sort_values(by='money', ascending=False)
df
```
